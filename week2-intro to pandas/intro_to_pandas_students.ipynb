# missing data interpolation
So now that we know how to look at missing data , let's find which values have missing vaalues . 

How do you deal with missing values ?????

1. Just remove them  : when is this good ? when is bad ?

2. Impute them , what does imputing means ? you can replace the value with mean or median or mode , if you have date data you might be able to replace the missing value with the last known value. . 

2. Impute them with the value of their nearest neighbor. 
3. In case of numerical value and linearity , use regression to predict the value.

4. Use algorithms that don't mind missing values.

5. Treat missing values as a class . This only works with categorical data or discreete data. 

6. Any other values come to your mind ?

How would you deal with missing value here ?
# which one is the best answer here ?


```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
#imp.fit_transform(np.ravel(df.random_val))
imp.fit(df['random_val'].values.reshape(-1, 1))
df['random_val'] = imp.transform(df['random_val'].values.reshape(-1, 1))
```
