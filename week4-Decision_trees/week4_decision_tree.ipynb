# Decision Trees
One of the most versatile Algorithm in Machine learning are decision Trees , The decision trees are the base later for lots of other learning algorithm such as random forest and XGboost.
So Let's start learn about them. 

Sklearn uses the CART(Classification And Regression Tree) Algortihm . The cart algorithm is a very simple algortihm that at every point it chooses a feature k , and a threshold tk to divide the data into to subset(left and right children. The pair of (k,tk) is chosen in a way that it maximizes the information gain or in other words find the purest subset at each step. The cart always split the data into two. However there are other decision tree algorithms that can split the data in several children.

$$ cost function = J(k, t_k) = \frac{m_{left}}{m} G_{left} + \frac{m_{right}}{m} G_{right} $$

where 
$ m_{left/right}$ : Calculates the number of intances in left right. 
and 
$G_{left/right}$ : is the measure of impurity 


# Gini and Entropy 
There are two measures that calculate the impurity/information gain. Gini index or Entropy. The default in sklearn is gini. 

We calculate each by the following formula : 

$$ Gini Index  = 1 - \sum_{i=1}^{n} p_i ^ 2 $$

and 

$$ Entropy = - \sum_{i=1}^{n} p_i * \log_2{p_i} $$

the two values are both between 0 and 1 and 0 meaning the purest and the 1 meaning the most randomness

# which one 

There usually don't make such a difference , however gini is slightly faster to calculate. 
they make a difference in the way tree is created:
gini tends to isolate the most frequent class in its own branch of the tree , as for entropy it tends to create more balanced trees.

# disadvantages
-  "Decision-tree learners can create over-complex trees that do not generalise the data well.(overfitting)
-  Decision tree learners create biased trees if some classes dominate. It is therefore recommended to balance the dataset prior to fitting with the decision tree.
- Decision trees can be unstable because small variations in the data might result in a completely different tree being generated. This problem is mitigated by using decision trees within an ensemble."sklearn
- Decision Trees love orthogonal decision boundaries (all splits are perpendicular to an axis), which makes them sensitive to training set rotation. 
- "More generally, the main issue with Decision Trees is that they are very sensitive to small variations in the training data."Geron

# The Bias/Variance Tradeoff

An important theoretical result of statistics and Machine Learning is the fact that a model‚Äôs generalization error can be expressed as the sum of three very different errors:
## Bias
this part of the generalization error is due to wrong assumptions, such as assum‚Äê ing that the data is linear when it is actually quadratic. A high-bias model is most likely to underfit the training data.10
## Variance
This part is due to the model‚Äôs excessive sensitivity to small variations in the training data. A model with many degrees of freedom (such as a high-degree pol‚Äê ynomial model) is likely to have high variance, and thus to overfit the training data.
## Irreducible error
This part is due to the noisiness of the data itself. The only way to reduce this part of the error is to clean up the data (e.g., fix the data sources, such as broken sensors, or detect and remove outliers).

## trade off
Increasing a model‚Äôs complexity will typically increase its variance and reduce its bias. Conversely, reducing a model‚Äôs complexity increases its bias and reduces its variance. This is why it is called a tradeoff.

- Selecting a more powerful model, with more parameters

overfitting fix :
- To simplify the model by selecting one with fewer parameters (e.g., a linear model rather than a high-degree polynomial model), by reducing the number of attributes in the training data or by constraining the model
